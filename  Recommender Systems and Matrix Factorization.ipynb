Recommenation as a matrix completion problem 

**Matrix completion** is the task of filling in the missing entries of a partially observed matrix. A wide range of datasets are naturally organized in matrix form. One example is the movie-ratings matrix, as appears in the Netflix problem: Given a ratings matrix in which each entry **(i, j)** represents the rating of movie **j** by customer **i** if customer **i** has watched movie 
***j*** and is otherwise missing, we would like to predict the remaining entries in order to make good recommendations to customers on what to watch next. 

Broadly, there are two types of recommendation systems: 

- **Content Based:** A content based recommender works with data that the user provides, either explicitly (rating) or implicitly (clicking on a link). Based on that data, a user profile is generated, which is then used to make suggestions to the user. As the user provides more inputs or takes actions on the recommendations, the engine becomes more and more accurate.

- **Collaborative filtering based:**Collaborative Filtering, on the other hand, doesn’t need anything else except users’ historical preference on a set of items. Because it’s based on historical data, the core assumption here is that the users who have agreed in the past tend to also agree in the future.

Memory-Based Collaborative Filtering approaches can be divided into two main sections: **user-item** filtering and **item-item** filtering.

**User-user:** find users that are similar to that user based on similarity of ratings, and recommend items that those similar users liked. 

**Item-Item:** find users who liked that item, and find other items that those users or similar users also liked.

**Matrix Factorization:** Division of matrix is such that if we multiply factorized matrix we will get original matrix.

PCA is a linear model in mapping d-dimensional input features to k-dimensional latent factors (k principal components).

PCA can interpreted example of matrix factorization,the PCA viewpoint requires that compute the eigenvalues and eigenvectors of the covariance matrix which is the product $C = XX^⊤/(n-1)$ , here X is the data matrix and C is covariance matrix. Since the covariance matrix is symmetric, the matrix is diagonalizable, and the eigenvectors can be normalized such that they are orthonormal: 

$C=WDW^⊤$ (where W is eigenvector and D is matrix of eigenvalues in diagonal elements and non-diagonal elements are zero).

**Singular value decomposition (SVD)** takes a rectangular matrix of gene expression data (defined as A, where A is a n x p matrix) in which the n rows represents the genes, and the p columns represents the experimental conditions. 

The SVD theorem states: 

$X_n*_p= U_n*_nS_n*_pV^T_p*_p$

Where
$U^TU = I_n*_n$ 

$V^TV = I_p*_p $ (i.e. U and V are orthogonal)

Where the columns of U are the left singular vectors (gene coefficient vectors); S (the same dimensions as A) has singular values and is diagonal (mode amplitudes); and VT has rows that are the right singular vectors (expression level vectors). The SVD represents an expansion of the original data in a coordinate system where the covariance matrix is diagonal.

Calculating the SVD consists of finding the eigenvalues and eigenvectors of $AA^T and  A^TA $. The eigenvectors of $A^TA$ make up the columns of V , the eigenvectors of $AA^T$ make up the columns of U. Also, the singular values in S are square roots of eigenvalues from $AA^T or A^TA$.  The singular values are the diagonal entries of the S matrix and are arranged in descending order. The singular values are always real numbers. If the matrix A is a real matrix, then U and V are also real.

On the other hand, applying SVD to the data matrix X as follows:

$A_n*_p= U_n*_nS_n*_pV^T_p*_p$

and attempting to construct the covariance matrix from this decomposition gives

$AA^⊤=(UΣV^⊤)(UΣV^⊤)^⊤ $ 

$AA^⊤=(UΣV^⊤)(VΣU^⊤) $ 

$C=A^⊤A/(n−1)$   covariance matrix C of PCA 

$ C=(UΣV^⊤)(VΣU^⊤)/(n−1) $

and since V is an orthogonal matrix $(V^⊤V=I)$,

$C=UΣ^2U^⊤/(n-1)$  

and the correspondence is easily seen (the square roots of the eigenvalues of $AA^⊤$ are the singular values of A, etc.) 

In fact, using the SVD to perform PCA makes much better sense numerically than forming the covariance matrix to begin with, since the formation of $XX^⊤$ can cause loss of precision.

**Non-Negative Matrix Factorization (NMF)** is a technique where a matrix  V  with nonnegative entries is factored into two matrices  W  and  H  that also have only nonnegative entries.

**Matrix Factorization for Collaborative filtering:** 
Firstly, we have a set of U users(n), and a set of I items (m). Let R be the matrix of size |n| x |m| that contains all the ratings that the users have assigned to the items.the input matrix (R) we have will be highly sparse as we know. Our task then, is to find two matrices, P (|n|x d) and Q (|m|x d) such that their product approximately equals to R is given by:

$ R ≈ PQ^T = \hat{R}$ prediction Rating 

To find the factor vectors ($p_n*_d and Q_m*_d$), the system minimizes the regularized squared error on the set of known ratings as (Set of non-missing entries) 

For the fitting procedure, we want every known $R_i,_j $ to be as close to $\hat{R}_i,_j =p^T_iq_j$ as possible. $p_i$ is the ith row of $P$. $q_j$ is the jth column of $Q$. 

For the prediction procedure, we won’t have an $R_i,_j $(sparse values), but we can use $\hat{R}_i,_j=p^T_iq_j$ to tell us what user i might rate movie j given the existing patterns. 

 $ argmin_(p,q) \sum_{\mathrm(i,j \in z)}(\hat{R}_i,_j -p^T_iq_j) $ (we can reslove this by using SGD) 

 Note: Where **Z** is the set of all pairs*(i, j)* where user *i* has rated movie *j*. 

we would have found those values of P and Q which will minimize the known values of R. At the same time, in the above process we are assuming that these values of P and Q will help us in predicting the unknown values of R (which are values in R that are not created by the users yet) as we are modifying all the values of P and Q to fit the known values of R

Based on these predicted values, we will recommend the movies to the user.

**Cold Start problem:**
Case 1: (When we have a new user)
In this case, we recommend top items based on the metadata of the user (like location of the user, the type of device being used by the user, the type of browser which is used for browsing the e-commerce site, etc). Those items purchased by the users who are from same location, using similar device(like ipad/iphone/android device/etc) are recommended to this new user. This will be purely content based recommender system until the user gives a decent and reasonable number of ratings for various items.

Case 2: (When a new item is added to the inventory)
In this case, information like which category does this item belong to(Like Electronics, Sports, clothing, books, etc), the cost of this item, description of this new item are taken into consideration and to those users who have bought items from same category, with similar cost and similar descriptions, this new item will be recommended. This is also a purely Content based recommender system until this new item gets a decent and reasonable number of ratings from various users.
